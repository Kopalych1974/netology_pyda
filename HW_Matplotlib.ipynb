### Визуализация данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('nhl_dataset')
data.head()
```

```python
nhl = data[['Salary', 'Born', 'City', 'Cntry', 'Nat', 'DftYr', 'DftRd', 'ENG', 'First Name', 'Last Name', 'FO%', 'G', 'Game', 'GP', 'PTS', 'GWG', 'Hand', 'Ht', 'Match', 'OTG', 'PIM', 'Position', 'Team', 'TOI', 'Wt', ]]
```

```python
nhl.head()
```

```python
rus_draft = nhl[nhl.Nat == 'RUS']
rus_draft
```

```python
draft_salary = rus_draft.groupby('DftYr').sum().sort_values('Salary', ascending=False)
```

```python
draft_salary.head()
```

```python
top_salary = nhl.sort_values(by='Salary', ascending=False).head(10)
```

```python
# для графика взял год проведения драфта и колличество русских, выбранных в этот год.
# для этого сгруппировал данные по году и посчитал количество значений
# при подготовке данных поставил фильтр на национальность русских в ДФ rus_draft
# график отражает в разрезе годов, количество выбранных русских

draft_rus = rus_draft.groupby('DftYr').count()
draft_rus.index = draft_rus.index.astype('int64')
plt.bar(draft_rus.index, draft_rus.Born, color='g')
plt.title('Выбор русских на драфтах')
plt.xlabel('Год драфта')
plt.ylabel('Выбрано русских')
plt.grid()  
```

```python
draft_salary = rus_draft.groupby('DftYr').sum()
draft_salary.index = draft_salary.index.astype('int64')
draft_salary
```

```python
# в график показывает для какого года драфта русских, было заключены наибольшие суммы контрактов

draft_salary.Salary.plot(kind='bar', rot=45)
plt.title('Самый высокооплачиваемый год драфта для русских')
plt.xlabel('Год драфта')
plt.ylabel('Зарплата')
```

```python
top_salary = nhl.sort_values(by='Salary').tail(10)
```

```python
# для этого графика отобрал самых высокооплачиваемых хоккеистов, по сумме контракта

plt.barh(top_salary['Last Name'], top_salary['Salary'])
plt.title('ТОП-10 крупных контрактов НХЛ')
plt.xlabel('Сумма контракта')
plt.ylabel('Фамилия игрока')
```

```python
nhl.groupby(['DftRd', 'Nat']).count().unstack().head()
```

```python
nat_draft = nhl[nhl.Nat.isin(['RUS', 'SWE', 'FIN', 'CZE', ])]
```

```python
data_draft = nat_draft.groupby(['DftRd', 'Nat']).count().unstack()['Salary']
```

```python
# отобрал европейцев и показал в каком раунде драфта, какое распределение национальностей
# русских забирают в первых трех раундах, слабых русских (после 3 раунда) команды не драфтуют

data_draft.plot(kind='bar', stacked=True)
plt.title('Выборы европейцев на драфте по раундам')
plt.xlabel('Количество выборов')
plt.ylabel('Раунд')
```

```python
right_vs_left = nhl[nhl.Position.isin(['C'])].groupby(by='Hand').mean()
```

```python
# сделал сравнение для центральных нападающих, кто лучше выигрывает взбрасывания с левым хватом или правым
# хват центрального не влияет на эффективность работы на взбрасывании

plt.bar(right_vs_left.index, right_vs_left['FO%'])
plt.title('Леворукие против праворуких')
plt.xlabel('Хват клюшки')
plt.ylabel('% выгранных вбрасываний')
```

```python
nhl['pts_per_game'] = nhl.PTS / nhl.GP
```

```python
points_draft = nhl[nhl.Position.isin(['LW', 'RW', 'C'])]
```

```python
# на графике отражено количество очков нибираемых за игру, для игроков выбранных в соответствующем раунде драфта

sns.boxplot(points_draft.DftRd, points_draft.pts_per_game)
```
